```python
import pandas as pd
import keras
```

```python
df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
df.head()
```

### A. Build a baseline model (5 marks)

Use the Keras library to build a neural network with the following:

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error as the loss function.

1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn.

```python
df_columns = df.columns
X = df[df_columns[df_columns != 'Strength']] # all columns except Strength
y = df['Strength'] # Strength column
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

2. Train the model on the training data using 50 epochs.

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_cols = X_train.shape[1]
```

```python
def reg_baseline_model():
    # create model
    model = Sequential()
    # One hidden layer of 10 nodes, and a ReLU activation function
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = reg_baseline_model()
```

```python
# fit the model
model.fit(X_train, y_train, epochs=50, verbose=3)
```

```python
y_pred = model.predict(X_test)
```

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE on test is equal to', mean_squared_error(y_test, y_pred))
```

3. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

```python
mse_50 = []
for i in range(0,50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    model.fit(X_train, y_train, epochs=50, verbose=3)
    y_pred = model.predict(X_test)
    mse_50.append( mean_squared_error(y_test, y_pred))
    print('MSE on test at running {} is equal to {}'.format(i, mean_squared_error(y_test, y_pred)))
```

5. Report the mean and the standard deviation of the mean squared errors.

```python
import numpy as np
```

```python
mean_mse_50 = np.mean(mse_50)
print(mean_mse_50)
```

```python
std_mse_50 = np.std(mse_50)
print(std_mse_50)
```
